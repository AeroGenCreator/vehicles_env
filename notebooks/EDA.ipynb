```python
'''Representando la frecuencia de la capacidad
de tanques a manera de histograma.'''
import pandas as pd
import plotly.graph_objects as go
# Lectura del DataFrame
df = pd.read_csv(r'D:\workspace_data_science\repositorios_clonados\vehicles_env\vehicles_us.csv')
# Establesco el valor -1 para los tanques desconocidos
df['cylinders'] = df['cylinders'].fillna(-1)
#Filtrando los tanques desconocidos:
df2 = df[~(df['cylinders'] == -1 )]
# Creacion del Histograma
Hist1 = go.Figure(data=[go.Histogram(x=df2['cylinders'],marker_color='skyblue',nbinsx=24)])
#Titulo:
Hist1.update_layout(
    title='Frecuencias en la capacidad de "Tanques"',
    title_x=0.5,
    xaxis_title='Capacidad del "Tanque"',
    yaxis_title='Cantidad de Unidades Vehiculares en "Miles"')
Hist1.show()
```

```python
'''Mostrando la correlacion entre
capacidad de tanques y gasto total'''
import pandas as pd
import plotly.graph_objects as go
# Lectura del DataFrame
df = pd.read_csv(r'D:\workspace_data_science\repositorios_clonados\vehicles_env\vehicles_us.csv')
# Establesco el valor -1 para los tanques desconocidos
df['cylinders'] = df['cylinders'].fillna(-1)
#Filtrando los tanques desconocidos:
df2 = df[~(df['cylinders'] == -1 )]
Scat1 = go.Figure(data=[go.Scatter(x=df2['odometer'],y=df2['price'], mode='markers', marker_color='orange')])
Scat1.update_layout(title_text= 'Relacion entre "Gasto Total" y "Capacidad de Tanques"')
Scat1.show()
```
